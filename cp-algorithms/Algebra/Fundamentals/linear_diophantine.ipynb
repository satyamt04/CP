# Linear Diophantine Equation

A linear Diophantine equation has integer solutions if and only if the greatest common divisor of $a$ and $b$ divides $c$.

For the equation:
$$ax + by = c$$

to have a solution, $c$ must be divisible by $\gcd(a, b)$:

$$c = k \cdot \gcd(a, b)$$

where $k$ is an integer.

```python
from math import gcd

def isPossible(a, b, c):
    return (c % gcd(a, b) == 0)

isPossible(4, 6, 8), isPossible(3, 6, 8)
```
